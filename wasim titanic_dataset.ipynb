```python
# importing and extrctng variables
titanic=pd.read_csv('titanic_dataset.csv')
X=titanic.iloc[: , :-1].values
y=titanic.iloc[: , :4].values
titanic.head()
```

```python
X=titanic['PassengerId'].to_numpy().reshape(-1,1)
y=titanic['Survived'].to_numpy().reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
plt.scatter(X_test,y_test,color="blue")
plt.plot(X_test,lr.predict(X_test),color="black")
plt.xlabel("PassengerID")
plt.ylabel("Survived")
plt.show()
```
